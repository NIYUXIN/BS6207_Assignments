```python
def pool2d(X, pool_size, mode='max'):
    X = X.float()
    p_h, p_w = pool_size
    Y = torch.zeros(X.shape[0],X.shape[1] - p_h + 1, X.shape[2] - p_w + 1)
    for i in range(Y.shape[0]):
        for j in range(Y.shape[1]):
            for k in range(Y.shape[2]):
                if mode == 'max':
                    Y[i, j, k] = X[i,j: j + p_h, k: k + p_w].max()
                elif mode == 'mean':
                    Y[i, j, k] = X[i,j: j + p_h, k: k + p_w].mean()       
    return Y
```

### 3 Conv2d

```python
def pool2d_one(X, weight,bias,channel):
    X = X.float()
    p_h=weight.shape[2]
    p_w=weight.shape[3]
    Y = torch.zeros(1,X.shape[1] - p_h + 1, X.shape[2] - p_w + 1)
    for i in range(X.shape[0]):
        for j in range(Y.shape[1]):
            for k in range(Y.shape[2]):
                    Y[0, j, k] += (X[i,j: j + p_h, k: k + p_w]*weight[channel][i]).sum()
    Y=Y+bias[channel]
    return Y

def poo2d_multi(X, weight,bias,out_channels):
    result=[]
    for i in range(out_channels):
        Y=pool2d_one(X, weight,bias,i)
        result.append(Y)
        
    return torch.stack(result,dim=1)
```

### 4 Conv2d

```python
def pool2d_one(X, weight,bias,channel,stride,dilation):
    X = X.float()
    p_h=weight.shape[2]
    p_w=weight.shape[3]
    Y = torch.zeros(1,(X.shape[1] - dilation*(p_h-1)-1)//stride + 1,
                    (X.shape[2] - dilation*(p_w-1)-1)//stride + 1)
    for i in range(X.shape[0]):
        for j in range(Y.shape[1],):
            for k in range(Y.shape[2]):

                candidate=X[i,j*stride: j*stride + p_h*dilation,
                                     k*stride: k*stride + p_w*dilation]
#                 print(candidate.shape)
                candidate_=torch.zeros(p_h,p_w)
                for m in range(p_h):
                    for n in range(p_w):
#                         print(n)
                        candidate_[m,n]=candidate[m*dilation,n*dilation]
                
                Y[0, j, k] += (candidate_*weight[channel][i]).sum()
    Y=Y+bias[channel]
    return Y

def poo2d_multi(X, weight,bias,out_channels,stride,dilation):
    result=[]
    for i in range(out_channels):
        Y=pool2d_one(X, weight,bias,i,stride,dilation)
        result.append(Y)
        
    return torch.stack(result,dim=1)
```

### 5 convtranspose

```python
def pool2d_one(X,weight,bias,channel):
    h= weight.shape[2]
    w=weight.shape[3]
    Y = torch.zeros((1,X.shape[1] + h - 1, X.shape[2] + w - 1))
#     print(Y.shape)
    for k in range(X.shape[0]):
        for i in range(X.shape[1]):
            for j in range(X.shape[2]):
                Y[0,i:i + h, j:j + w] += X[k,i, j] * weight[k][channel]
    Y+=bias[channel]
    return Y
def poo2d_multi(X, weight,bias,out_channels):
    result=[]
    for i in range(out_channels):
#         print(i)
        Y=pool2d_one(X, weight,bias,i)
#         print(Y.shape)
        result.append(Y)
        
    return torch.stack(result,dim=1)
```

```python
def norm(X):
    eps=1e-05
    me=running_mean.view(3,1,1)
    var=running_var.view(3,1,1)
    return (X-me)/torch.sqrt(var+eps)
```

```python
#inp,1,C,H,W, target: 1,H,W
class cross_entropy:
    def __init__(self,inp,target):
        self.inp=inp
        self.target=target
        self.H=target.shape[1]
        self.W=target.shape[2]
    def forward(self):
        return self.nn_loss(self.log_softmax())
    def log_softmax(self):
        exp=torch.exp(self.inp[0])
        sum_=torch.sum(exp,dim=0)
        soft_max=exp/sum_
        return torch.log(soft_max)
    def nn_loss(self,inp):
        loss=0
        for i in range(self.H):
            for j in range(self.W):
                index=target[0,i,j]
                num=inp[index,i,j]
                loss-=num
        return (loss)/self.H/self.W
```

```python
cs=cross_entropy(inp,target)
cs.forward()
```

```python
result_s["cross_entropy"]=np.array([cs.forward().detach().numpy()]).reshape(1,-1)
```

```python
def save_file(List_file,path):
    with open(path,"w") as f:
        for k,v in List_file.items():
            f.write(k+': \n')
            for i in range(v.shape[0]):
                np.savetxt(f,List_file[k][i])
                f.write('\n')
            f.write('\n')
            f.write('_______________________'+'\n')
    f.close()
```
